```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('monkeypox_df.csv')
```

```python
df.head(20)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df=df.drop(['ID','Hospitalised (Y/N/NA)','Isolated (Y/N/NA)','Travel_history (Y/N/NA)','Date_onset','Date_confirmation','Date_hospitalisation','Date_isolation','Outcome','Contact_comment','Travel_history_location','Contact_ID','Travel_history_entry','Travel_history_entry','Travel_history_start','Travel_history_country','Contact_location','Genomics_Metadata','Confirmation_method','Source','Source_II','Date_entry','Date_last_modified','Source_III','Country_ISO3'], axis = 1)
```

```python
df.shape
```

```python
df
```

```python
df.isnull().sum()
```

```python
ndf = df.dropna()
```

```python
ndf
```

```python
ndf.shape
```

```python
x = ndf.drop(['Status'], axis = 1)
```

```python
x
```

```python
y = ndf['Status']
```

```python
y
```

```python
from sklearn import preprocessing
le_x= preprocessing.LabelEncoder()
```

```python
x.Location = le_x.fit_transform(x.Location)
x.City = le_x.fit_transform(x.City)
x.Country = le_x.fit_transform(x.Country)
x.Age = le_x.fit_transform(x.Age)
x.Gender = le_x.fit_transform(x.Gender)
x.Symptoms = le_x.fit_transform(x.Symptoms)
y = le_x.fit_transform(y)

```

```python
x
```

```python
x.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=1)
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
reg.fit(x_train,y_train)
```

```python
prediction = reg.predict(x_test)
```

```python
prediction
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,prediction))
```

```python
from sklearn.metrics import  confusion_matrix
```

```python
print(confusion_matrix(y_test, prediction))
```

**GaussianNB**

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
pred = gnb.predict(x_test)
```

```python
pred
```

```python
gnb.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

```python
pree = dtc.predict(x_test)
```

```python
pree
```

```python
dtc.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pree))
```
